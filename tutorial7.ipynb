### List Comprehensions

```python
li = [3, 6, 2, 7]
out = []
for elem in li:
    print(out)
    out.append(elem*12)
print(out)
```

```python
[elem*12 for elem in li]#자주 사용
```

```python
li = [('a', 1), ('b', 2), ('c', 7)]
[ n * 3 for (x, n) in li]#코드 빈칸 채우기
```

```python
li = [3, 6, 2, 7, 1, 9]
[elem *2 for elem in li if elem >4]#자주 사용
```

```python
li = [3, 2, 4, 1]
[elem*2 for elem in 
    [item+1 for item in li] ]##시험 나올 수도
```
